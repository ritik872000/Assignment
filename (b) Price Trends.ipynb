```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df=pd.read_csv(r'C:\Users\ritik\Desktop\Gramoday\agri_data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#converting obj to datetime
df['Price Date'] = pd.to_datetime(df['Price Date'])
```

```python
df.head()
```

```python
# creaitng the new column for months, quarter
df['Month']=pd.to_datetime(df['Price Date'] ).dt.month
# creaitng the new column for quarter
df['Quarter']=pd.to_datetime(df['Price Date'] ).dt.quarter
```

```python
# 
df.groupby('Market Name')['Modal Price (Rs./Quintal)'].count()
```

We can clearly see that <b>Achnera</b> is the major market 

```python
df.head()
```

```python
# creating seperate dataframes for different markets
Achnera=df.loc[df['Market Name'] == 'Achnera']
Agra=df.loc[df['Market Name'] == 'Agra']
Fatehabad=df.loc[df['Market Name'] == 'Fatehabad']
Fatehpur_Sikri=df.loc[df['Market Name'] == 'Fatehpur Sikri']
Jagnair=df.loc[df['Market Name'] == 'Jagnair']
Jarar=df.loc[df['Market Name'] == 'Jarar']
Khairagarh=df.loc[df['Market Name'] == 'Khairagarh']
Samsabad=df.loc[df['Market Name'] == 'Samsabad']
```

# Yearly Analysis

```python
# plotting date vs prices for each market
plt.figure(figsize=(15,8))

plt.plot(Achnera['Price Date'], Achnera['Modal Price (Rs./Quintal)'], label='Achnera',color='blue')
plt.plot(Agra['Price Date'], Agra['Modal Price (Rs./Quintal)'],label='Agra',color='red')
plt.plot(Fatehabad['Price Date'], Fatehabad['Modal Price (Rs./Quintal)'],label='Fatehabad',color='green')
plt.plot(Fatehpur_Sikri['Price Date'], Fatehpur_Sikri['Modal Price (Rs./Quintal)'],label='Fatehpur_Sikri',color='magenta')
plt.plot(Jagnair['Price Date'], Jagnair['Modal Price (Rs./Quintal)'],label='Jagnair',color='yellow')
plt.plot(Jarar['Price Date'], Jarar['Modal Price (Rs./Quintal)'],label='Jarar',color='cyan')
plt.plot(Khairagarh['Price Date'], Khairagarh['Modal Price (Rs./Quintal)'],label='Khairagarh',color='black')
plt.plot(Samsabad['Price Date'], Samsabad['Modal Price (Rs./Quintal)'],label='Samsabad',color='orange')

plt.legend(loc='upper left')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Date VS Price')
plt.show()
```

We can notice that prices mostly increase till the month of november and suddenly drops in the last month.

Probable reasons are :

1. In december supply increase due to which price falls.
2. There are many factors behind the increase in prices(Weather, Soil condition).

# Quarterly Analysis

```python
# plotting date vs prices for each market
plt.figure(figsize=(15,8))

plt.plot(Achnera['Quarter'], Achnera['Modal Price (Rs./Quintal)'], label='Achnera',color='blue')
plt.plot(Agra['Quarter'], Agra['Modal Price (Rs./Quintal)'],label='Agra',color='red')
plt.plot(Fatehabad['Quarter'], Fatehabad['Modal Price (Rs./Quintal)'],label='Fatehabad',color='green')
plt.plot(Fatehpur_Sikri['Quarter'], Fatehpur_Sikri['Modal Price (Rs./Quintal)'],label='Fatehpur_Sikri',color='magenta')
plt.plot(Jagnair['Quarter'], Jagnair['Modal Price (Rs./Quintal)'],label='Jagnair',color='yellow')
plt.plot(Jarar['Quarter'], Jarar['Modal Price (Rs./Quintal)'],label='Jarar',color='cyan')
plt.plot(Khairagarh['Quarter'], Khairagarh['Modal Price (Rs./Quintal)'],label='Khairagarh',color='black')
plt.plot(Samsabad['Quarter'], Samsabad['Modal Price (Rs./Quintal)'],label='Samsabad',color='orange')

plt.legend(loc='upper left')
plt.xticks([1,2,3,4])
plt.xlabel('Quarter')
plt.ylabel('Price')
plt.title('Date VS Price')
plt.show()
```

If we see quarter wise theres <b>not much difference in 1st and 2nd quarter</b> but theres a <b>substantial difference in 3rd quater</b> i.e

During the months of July, August and September prices are the highest
